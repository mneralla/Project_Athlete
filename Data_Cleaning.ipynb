```python
import os
import pandas as pd
import json
import csv
```

```python
athletes_df = pd.read_csv("Forbes Richest Atheletes (Forbes Richest Athletes 1990-2019).csv")
countries_df = pd.read_csv("countries.csv")
```

```python
countries_df.head()
```

```python
countries_df.rename(columns={'name':'Nationality'}, inplace=True)
```

```python
countries_df.head()
```

```python
athletes_df['Nationality'] = athletes_df['Nationality'].replace(
    {'USA': 'United States', 'Dominican': 'Dominican Republic', 'Northern Ireland': 'Ireland', 'Filipino': 'Philippines', 'UK': 'United Kingdom'})
```

```python
athletes_df.head()
```

```python
merge_table = pd.merge(athletes_df, countries_df, on="Nationality", how="left")
merge_table
```

```python
merge_table.count()
```

```python
merge_table.to_csv('athletes_merged.csv')
```
